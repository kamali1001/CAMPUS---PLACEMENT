```python
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(random_state=42, n_jobs=-1, max_depth=5, n_estimators=100, oob_score=True)
regressor.fit(X_train, Y_train) 
```

```python
dataset.head()
place_predict = regressor.fit(X_train, Y_train)
```

```python
place_predict.predict([[100,100,100,100,100,100,100,100,100,100,100,100,100,100,100]])
```

```python
import pickle as pk
file = pk.dump(place_predict, open('place_predict.pkl', 'wb'))
```
